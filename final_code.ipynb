```python
import numpy as np

# Function to handle division and zero-check
def safe_divide(num, denom):
    if denom == 0:
        return 1 if num == 0 else 2
    return num / denom

# Calculate avg_st_rt_ratio
train_data['avg_st_rt_ratio'] = train_data.apply(
    lambda row: safe_divide(row['Average Strike Rate of Top 5 for team 1'], row['Average Strike Rate of Top 5 for team 2']),
    axis=1
)

# Calculate head_to_head_performance for batting average
train_data['avg_bat_avg_ratio'] = train_data.apply(
    lambda row: safe_divide(row['Average Batting Average of Top 5 for team 1'], row['Average Batting Average of Top 5 for team 2']),
    axis=1
)

# Calculate head_to_head_performance for economy rate
train_data['avg_eco_bowl_ratio'] = train_data.apply(
    lambda row: safe_divide(row['Average Economy of Bottom 5 Bowlers for team 1'], row['Average Economy of Bottom 5 Bowlers for team 2']),
    axis=1
)

# Calculate head_to_head_performance for bowling strike rate
train_data['avg_bowl_st_rt_ratio'] = train_data.apply(
    lambda row: safe_divide(row['Average Strike Rate of Bottom 5 Bowlers for team 1'], row['Average Strike Rate of Bottom 5 Bowlers for team 2']),
    axis=1
)

# Calculate the ultimate score
train_data['ultimate_score'] = (
    train_data[['random', 'scoreinf', 'score_ratio', 'final_day', 'achaa']].sum(axis=1) / 5
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import VotingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import (
    accuracy_score, confusion_matrix, roc_curve, auc, classification_report, 
    precision_recall_curve, average_precision_score
)
from sklearn.calibration import calibration_curve
from sklearn.model_selection import learning_curve

# Define parameter values for XGBoost and LightGBM
xgb_params = {
    'learning_rate': 0.199,
    'n_estimators': 51,
    'max_depth': 3,
    'min_child_weight': 5,
    'subsample': 0.4,
    'colsample_bytree': 0.4,
    'reg_alpha': 1,
    'reg_lambda': 5,
    'objective': 'binary:logistic',
    'n_jobs': 4,
    'random_state': 28,
    'use_label_encoder': False,
    'eval_metric': 'logloss'
}

lgb_params = {
    'learning_rate': 0.198,
    'n_estimators': 54,
    'max_depth': 3,
    'min_child_weight': 3,
    'subsample': 0.4,
    'colsample_bytree': 1.0,
    'reg_alpha': 2,
    'reg_lambda': 3,
    'objective': 'binary',
    'n_jobs': 4,
    'random_state': 28
}

# Initialize the models with the provided parameters
xgb_model = XGBClassifier(**xgb_params)
lgb_model = LGBMClassifier(**lgb_params)

# Fit the models using all the data
xgb_model.fit(X, y)
lgb_model.fit(X, y)

# Create a Voting Classifier with the fitted models
clf_gbm = VotingClassifier(
    estimators=[
        ('xgb', xgb_model),
        ('lgb', lgb_model)
    ],
    voting='soft'  # Soft voting to consider the predicted probabilities
)

# Fit the ensemble model using all the data
clf_gbm.fit(X, y)

# Since there's no validation set, we'll use the training data itself for prediction and evaluation
y_pred = clf_gbm.predict(X)
accuracy = accuracy_score(y, y_pred)
print(f'Ensemble model accuracy: {accuracy:.4f}')

train_data['y_pred_01'] = clf_gbm.predict(X)
test_data['y_pred_01'] = clf_gbm.predict(X_test)

# Print classification report
print(classification_report(y, clf_gbm.predict(X), labels=[0, 1]))

# Confusion Matrix
cm = confusion_matrix(y, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[0, 1], yticklabels=[0, 1])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# ROC Curve
y_prob = clf_gbm.predict_proba(X)[:, 1]
fpr, tpr, thresholds = roc_curve(y, y_prob)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()

# Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y, y_prob)
average_precision = average_precision_score(y, y_prob)
plt.figure()
plt.step(recall, precision, where='post', color='b', alpha=0.2, linestyle='-', linewidth=2)
plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve: AP={average_precision:.2f}')
plt.show()

# Calibration Curve
prob_true, prob_pred = calibration_curve(y, y_prob, n_bins=10)
plt.figure()
plt.plot(prob_pred, prob_true, 's-', label='Calibrated model')
plt.plot([0, 1], [0, 1], 'k--', label='Perfectly calibrated')
plt.xlabel('Mean predicted probability')
plt.ylabel('Fraction of positives')
plt.title('Calibration Curve')
plt.legend(loc='best')
plt.show()

# Residual Plot
residuals = y - y_pred
plt.figure()
plt.hist(residuals, bins=50, alpha=0.5, color='g', edgecolor='black')
plt.xlabel('Residual')
plt.ylabel('Frequency')
plt.title('Residuals Histogram')
plt.show()
```
